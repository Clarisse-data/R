```r
# Taking input from user: rain status, temperature in Celsius, and wind speed in km/hr to suggest activities

temperature_input <- readline(prompt= "Enter today's temperature (Celsius): ")
temperature <- as.numeric(temperature_input)

wind_input <- readline(prompt= "Enter wind speed (km/hr): ")
wind <- as.numeric(wind_input)

rain_input <- readline(prompt= "Is it raining (yes/no): ")
rain <- ifelse(tolower(rain_input) == "yes", TRUE, FALSE)

suggestion <- ""

# If it is not raining
if (!rain) {
  if (temperature >= 35) {
    if (wind <= 10) {
      suggestion <- "It's a very hot day with little wind. Consider swimming, going to the beach, or staying indoors with air conditioning."
    } else {
      suggestion <- "It's a very hot and windy day. Indoor activities with air conditioning are recommended."
    }
  } else if (temperature >= 20) {
    if (wind <= 15) {
      suggestion <- "It's warm and calm. Ideal for outdoor activities like cycling, hiking, or walking."
    } else {
      suggestion <- "It's warm but windy. Wear a light jacket and enjoy outdoor activities like hiking or a walk."
    }
  } else if (temperature >= 5) {
    if (wind <= 10) {
      suggestion <- "It's cool but calm. A light walk or some indoor activities would be a good choice."
    } else {
      suggestion <- "It's cool and windy. Prefer indoor activities or dress warmly for a short walk."
    }
  } else {
    if (wind <= 10) {
      suggestion <- "It's cold and calm. Indoor activities or a brief walk outside are recommended."
    } else {
      suggestion <- "It's cold and windy. Stay indoors and keep warm."
    }
  }

} else { # If it is raining
  if (temperature >= 35) {
    if (wind <= 10) {
      suggestion <- "It's a hot and rainy day with little wind. If you go outside, consider beach or pool activities with rain gear."
    } else {
      suggestion <- "It's hot, rainy, and windy. Best to stay indoors and avoid outdoor activities."
    }
  } else if (temperature >= 20) {
    if (wind <= 15) {
      suggestion <- "It's warm and rainy. Perfect for a raincoat walk or indoor activities."
    } else {
      suggestion <- "It's warm, rainy, and windy. A raincoat and light jacket are needed if venturing outside, or consider staying indoors."
    }
  } else if (temperature >= 5) {
    if (wind <= 10) {
      suggestion <- "It's cool and rainy. Indoor activities are preferable."
    } else {
      suggestion <- "It's cool, rainy, and windy. Stay indoors to avoid the harsh weather."
    }
  } else {
    if (wind <= 10) {
      suggestion <- "It's cold and rainy. Indoor activities with a warm drink are recommended."
    } else {
      suggestion <- "It's cold, rainy, and windy. Stay indoors and keep warm."
    }
  }
}

print(paste("Suggestions: ", suggestion))
```
